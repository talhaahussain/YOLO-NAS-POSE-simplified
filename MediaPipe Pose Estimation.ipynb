# Human Pose Estimation with MediaPipe and Python

This short project aims to use MediaPipe and OpenCV for HPE, to track the number of bicep curl reps performed by a user at the gym.

This is a precursor to the ECM3401 Continuous Assessment project, and exists to allow me to familiarise myself with HPE techniques and using the Google Colab platform.

The directions for this project are followed from Nicholas Renotte; see video <a href="https://youtu.be/06TE_U21FK4">here</a>.
